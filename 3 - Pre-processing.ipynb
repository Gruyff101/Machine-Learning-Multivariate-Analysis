```python
import pandas as pd

# Import
pd.set_option("display.max_columns", None)

df = pd.read_csv("RawJGM.csv", na_values=" ")

df
```

```python
# Check NaNs per column
pd.DataFrame(df.isna().sum()).T
```

```python
df.isna().sum().sum()
```

```python
# Drops columns that are not useful or cause data leakage
df = df.drop(columns=["team_nr", "Organisatie", "tijd", "filter_$"])

df.isna().sum().sum()
```

```python
# Drop all columns with more than 250 (~50%) of NaNs
# Reason we drop these is because there isn't enough data in the column to accurately impute
df = df.loc[:, df.isnull().sum() < 250]

df.isna().sum().sum()
```

```python
pd.DataFrame(df.isna().sum()).T
```

```python
# Drop rows with features which are NaN which aren't really inputable
df = df.dropna(subset=["aantalteam", "leeftijd_gem", "geslacht", "ervaring", "gehaald"])
```

```python
# Map to binary
df["gehaald"] = df["gehaald"].map({1: 0, 2: 1})
```

```python
df
```

```python
# Remove rows that contain more than 50% NaNs.
# Reason we drop these is because there isn't enough data in the row to accurately impute
threshold = len(df.columns) * 0.5
df = df[df.isna().sum(axis=1) <= threshold]

df
```

```python
from sklearn.impute import KNNImputer

# Impute remaining NaNs using KNN
imputer = KNNImputer()

df = pd.DataFrame(imputer.fit_transform(df), columns=df.columns)

df.isna().sum().sum()
```

```python
# Transform float dtypes to int
df = df.astype({"aantalteam": int, "gehaald": int})

df
```

```python
# Export
df.to_csv("JGM_new.csv", index=False)
```
